```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import zscore, ttest_ind
from scipy.fft import fft, fftfreq
from scipy.optimize import curve_fit

# Đọc dữ liệu và lấy 1008 điểm đầu tiên (7 ngày)
data = pd.read_csv('/content/Data.csv')
temperature = data['Temperature'].values[:1008]

# Tiền xử lý dữ liệu
def preprocess_data(temperature):
    temperature = pd.Series(temperature).interpolate().values if np.any(np.isnan(temperature)) else temperature
    temperature = temperature * 1.3  # Hiệu chỉnh nhiệt độ
    temperature_smooth = pd.Series(temperature).rolling(window=15, center=True).mean().values  # Tăng cửa sổ lọc
    return np.nan_to_num(temperature_smooth, nan=np.mean(temperature))

temperature_processed = preprocess_data(temperature)
delta_t = 10 / 60  # 10 phút = 1/6 giờ
times = np.arange(0, len(temperature_processed) * delta_t, delta_t)

# 1. FFT: Biểu diễn phổ và xác định chu kỳ chính
def fft_analysis(temperature, delta_t):
    n = len(temperature)
    fft_vals = fft(temperature)
    fft_magnitude = 2 * np.abs(fft_vals)[:n//2] / n
    freqs = fftfreq(n, d=delta_t)[:n//2]
    periods = 1 / freqs

    # Lọc chu kỳ hợp lý (0.5–12 giờ, tập trung sinh học)
    valid_indices = (periods >= 0.5) & (periods <= 12) & (freqs > 1/24)
    periods, fft_magnitude = periods[valid_indices], fft_magnitude[valid_indices]
    sort_indices = np.argsort(periods)
    periods, fft_magnitude = periods[sort_indices], fft_magnitude[sort_indices]

    # Xác định chu kỳ chi phối
    threshold = np.mean(fft_magnitude) + 1.5 * np.std(fft_magnitude)  # Giảm ngưỡng để nhạy hơn
    dominant_periods = periods[fft_magnitude > threshold]

    plt.figure(figsize=(8, 4))
    plt.plot(periods, fft_magnitude, label='Phổ tần số')
    plt.axvline(x=24, color='r', linestyle='--', label='Chu kỳ 24 giờ (tham chiếu)')
    plt.xlabel('Chu kỳ (giờ)')
    plt.ylabel('Biên độ')
    plt.title('Phổ tần số nhiệt độ cơ thể (FFT)')
    plt.legend()
    plt.grid()
    plt.show()

    return dominant_periods

# 2. Trung bình theo giờ: So sánh ngày/đêm
def hourly_comparison(temperature):
    samples_per_hour = 6
    samples_per_day = 24 * samples_per_hour

    # Tính trung bình theo giờ qua 7 ngày
    hourly_data = np.array([np.mean(temperature[i:i + samples_per_hour])
                           for i in range(0, len(temperature), samples_per_hour)])
    hourly_data = hourly_data.reshape(7, 24)
    hourly_avg = np.mean(hourly_data, axis=0)
    hours = np.arange(0, 24, 1)

    # Phân chia ngày (7:00–19:00) và đêm (19:00–7:00)
    day_temps = hourly_data[:, 7:19].flatten()
    night_temps = np.concatenate([hourly_data[:, :7], hourly_data[:, 19:]], axis=1).flatten()

    day_mean, night_mean = np.mean(day_temps), np.mean(night_temps)
    day_std, night_std = np.std(day_temps), np.std(night_temps)
    p_value = ttest_ind(day_temps, night_temps).pvalue

    plt.figure(figsize=(8, 4))
    plt.plot(hours, hourly_avg, label='Nhiệt độ trung bình/giờ', marker='o')
    plt.axvspan(7, 19, color='blue', alpha=0.1, label='Ngày (7:00–19:00)')
    plt.axvspan(0, 7, color='orange', alpha=0.1, label='Đêm (0:00–7:00)')
    plt.axvspan(19, 24, color='orange', alpha=0.1)
    plt.axhline(y=day_mean, color='b', linestyle='--', label=f'Ngày: {day_mean:.2f}°C')
    plt.axhline(y=night_mean, color='orange', linestyle='--', label=f'Đêm: {night_mean:.2f}°C')
    plt.xlabel('Giờ')
    plt.ylabel('Nhiệt độ (°C)')
    plt.title('So sánh nhiệt độ ngày/đêm (trung bình 7 ngày)')
    plt.legend()
    plt.grid()
    plt.show()

    return day_mean, day_std, night_mean, night_std, p_value

# 3. Cosine Fitting + Z-score: Nhận diện rối loạn
def cosine_zscore_analysis(temperature, times):
    def cosine_func(t, A, phi, offset):
        period = 24
        return A * np.cos(2 * np.pi * t / period + phi) + offset

    p0 = [np.std(temperature), 0, np.mean(temperature)]
    popt, _ = curve_fit(cosine_func, times, temperature, p0=p0)
    fitted = cosine_func(times, *popt)

    residuals = temperature - fitted
    z_scores = zscore(residuals)
    anomalies = np.where(np.abs(z_scores) > 2)[0]  # Giảm ngưỡng xuống 2

    circadian_amplitude = abs(popt[0])
    circadian_phase = popt[1] * 24 / (2 * np.pi) % 24  # Pha trong 24 giờ
    circadian_stable = circadian_amplitude > 0.5

    plt.figure(figsize=(10, 5))
    plt.subplot(2, 1, 1)
    plt.plot(times, temperature, label='Nhiệt độ thực tế', alpha=0.6)
    plt.plot(times, fitted, label='Mô hình cosin (24 giờ)', linestyle='--')
    plt.scatter(times[anomalies], temperature[anomalies], color='r', label='Bất thường')
    plt.xlabel('Thời gian (giờ)')
    plt.ylabel('Nhiệt độ (°C)')
    plt.title('Mô hình cosin và phát hiện bất thường')
    plt.legend()
    plt.grid()

    plt.subplot(2, 1, 2)
    plt.plot(times, residuals, label='Độ lệch (residuals)', alpha=0.6)
    plt.axhline(y=0, color='gray', linestyle='--')
    plt.scatter(times[anomalies], residuals[anomalies], color='r', label='Bất thường')
    plt.xlabel('Thời gian (giờ)')
    plt.ylabel('Độ lệch (°C)')
    plt.title('Độ lệch so với nhịp sinh học 24 giờ')
    plt.legend()
    plt.grid()
    plt.tight_layout()
    plt.show()

    print("\nNhận diện rối loạn nhịp sinh học:")
    print(f"- Biên độ chu kỳ 24 giờ: {circadian_amplitude:.2f}°C")
    print(f"- Pha chu kỳ 24 giờ: {circadian_phase:.2f} giờ (đỉnh nhiệt độ)")
    if len(anomalies) > 0:
        print(f"- Có {len(anomalies)} điểm bất thường → Có thể rối loạn.")
    if not circadian_stable:
        print("- Biên độ chu kỳ 24 giờ thấp → Có khả năng rối loạn.")
    else:
        print("- Nhịp sinh học 24 giờ ổn định.")

# Chạy phân tích
print("1. FFT: Phổ và chu kỳ chính")
dominant_periods = fft_analysis(temperature_processed, delta_t)
print("Chu kỳ chi phối (giờ):", [f"{p:.2f}" for p in dominant_periods])

print("\n2. Trung bình theo giờ: So sánh ngày/đêm")
day_mean, day_std, night_mean, night_std, p_value = hourly_comparison(temperature_processed)
print(f"- Ngày: {day_mean:.2f}±{day_std:.2f}°C")
print(f"- Đêm: {night_mean:.2f}±{night_std:.2f}°C")
print(f"- p-value: {p_value:.4f}")

print("\n3. Cosine Fitting + Z-score: Nhận diện rối loạn")
cosine_zscore_analysis(temperature_processed, times)
```
